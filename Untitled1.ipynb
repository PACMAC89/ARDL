###Looking for the ticker you need and import it

```python
%pip install yfinance
import yfinance as yf
ticker ="MSFT"

df_MSFT = yf.download(tickers = ticker)
```

###Visualize the data with plotly. Copy the code snippet from the web site https://plotly.com/python/candlestick-charts/

look at the data frame. man sieht date ist hier keie column sondern ein index. daher muss der code angepasst werrden

```python
df_MSFT
```

```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime

df = df_MSFT

#df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')

fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```
